# Edge Detection Technique

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('original_setsun.jpg')
img = cv2.cvtColor(img ,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
image = cv2.imread("original_setsun.jpg")
edges = cv2.Canny(image,50,300)
cv2.imwrite('original_setsun.jpg',edges)
plt.imshow(image)
```
