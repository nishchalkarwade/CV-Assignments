# Image Transformation

# ORIGINAL IMAGE

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('lens.jpg')
img = cv2.cvtColor(img ,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
print('number of pixels:', img.size)
print('dimension:',img.shape)
```

# Image Resizing or Scaling

```python
width,height = 200,200
img_resized = cv2.resize(img,(width,height))
plt.imshow(img_resized)
plt.xticks([]),plt.yticks([])
plt.show()
```

# Image Translation

```python
num_rows, num_cols = img.shape[:2]
translation_matrix = np.float32([ [1,0,100], [0,1,110] ])
image_translation = cv2.warpAffine(img, translation_matrix, (num_cols, num_rows))
plt.imshow(image_translation)
plt.xticks([]),plt.yticks([])
plt.show()
```

# Image Rotation

```python
height = img.shape[0]
width = img.shape[1]
def rotate_img(image, angle, rotating_points=None):
    if rotating_points is None:
        #rotate from center
        rotating_points = (width//2, height//2)

    dimensions =(width,height)
    rotateMatx = cv2.getRotationMatrix2D(rotating_points, angle, 1.0)  #1.0 is scaling size
    return cv2.warpAffine(img, rotateMatx,dimensions)
rotated_image = rotate_img(img, 45)  #rotate image by 45 degree
plt.imshow(rotated_image)
plt.xticks([]),plt.yticks([])
plt.show()
```
