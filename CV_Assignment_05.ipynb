# Image Thresholding Techniques

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('x.jpg',50)
```

```python
ret,thr1 = cv.threshold(img,110,255,cv.THRESH_BINARY)
ret,thr2 = cv.threshold(img,110,255,cv.THRESH_BINARY_INV)
ret,thr3 = cv.threshold(img,110,255,cv.THRESH_TRUNC)
ret,thr4 = cv.threshold(img,110,255,cv.THRESH_TOZERO)
ret,thr5 = cv.threshold(img,110,255,cv.THRESH_TOZERO_INV)
```

```python
#TITLES FOR IMAGES
titles = ['Original Image','BINARY','BINARY INVERTED','TRUNCATED','TOZERO','TOZERO_INVERTED']
images = [img, thr1, thr2, thr3, thr4, thr5]
```

```python
# PLOTTING SIDE BY SIDE
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
